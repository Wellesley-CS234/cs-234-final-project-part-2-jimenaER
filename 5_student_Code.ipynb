```python
from nltk.corpus import stopwords
```

```python
#Filter stop words fuction
def removeStop(text):
    try:

        nltk_stopwords_eng = set(stopwords.words('english'))
        words= word_tokenize(text)
        filtered_words=[word.lower() for word in words if word.lower() not in nltk_stopwords_eng]
        if len(filtered_words)< len(words):
            return ' '.join(filtered_words)
    except LookupError:
        pass

    try:
        nltk_stopwords_span = set(stopwords.words('spanish'))
        words= word_tokenize(text)
        filtered_words=[word.lower() for word in words if word.lower() not in nltk_stopwords_span]
        return ' '.join(filtered_words)
    except LookupError:
        return text
```

```python
#lowercase and Remove punctuations
def removePunctuation(text):
    punctuation= string.punctuation

    text_lower=text.lower()
    text_unpunctuated = text_lower.translate(str.maketrans({p: ' ' for p in string.punctuation}))

    return text_unpunctuated.strip()
    
```

```python
#cleaning classify_text by removing punctuation
p3_f1_data['classify_text']=p3_f1_data['classify_text'].apply(removePunctuation)
p3_f1_data['classify_text']=p3_f1_data['classify_text'].apply(removeStop)
p3_f1_data
```

# Categorize: Zero shot classification

```python
import pandas as pd
```

```python
fig1_data=pd.read_csv('student_csv/figure1_cleaned_data.csv')
```

```python
#This is the catrgories for both our zero shot classifications: S:Sports, NS:No Sports
labels=['Deportes','No deportes']
```

```python
import pandas as pd
from transformers import pipeline
from tqdm.notebook import tqdm
import torch
```

```python
#Start classifier
MODEL_NAME = "microsoft/xtremedistil-l6-h384-uncased"
DEVICE = 0 if torch.cuda.is_available() else -1

print(f"Loading model: {MODEL_NAME} on device: {'GPU' if DEVICE == 0 else 'CPU'}")

# pipeline is a function from HuggingFace's transformers library
classifier = pipeline(
    "zero-shot-classification",
    model=MODEL_NAME,
    device=DEVICE,
    batch_size=32
)

classifier
```

```python
#Use classifier to label based on title
resultsDE = classifier(
        fig1_data['classify_text'].to_list(),
        candidate_labels=labels,
        hypothesis_template= "This text is a {}.",
        multi_label=False
    )

resultsDE
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
def plot_confusion_matrix_with_percentages(y_true, y_pred, 
                                           labels=['Not Sports', 'Sports'], 
                                           title="Confusion Matrix"):
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    cm_sum = cm.sum()
    cm_percent = cm / cm.sum(axis=1, keepdims=True) * 100

    annot = [[f"{count}\n{pct:.1f}%" for count, pct in zip(row, row_pct)] 
             for row, row_pct in zip(cm, cm_percent)]

    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=annot, fmt="", cmap="Blues", 
                xticklabels=labels, yticklabels=labels)
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(title)
    plt.show()
```

```python
yinput=fig1_data['categories_generated'].astype(str).to_list()[20:30]
```
